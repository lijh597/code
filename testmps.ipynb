```python
import torch

# 1. 检查 MPS (Metal Performance Shaders) 是否可用
if torch.backends.mps.is_available():
    # 2. 如果可用，设置设备为 'mps'
    device = torch.device("mps")
    print("✅ MPS (GPU) is available. Using 'mps' device.")
else:
    # 3. 如果不可用，退回到 CPU
    device = torch.device("cpu")
    print("⚠️ MPS (GPU) not available. Using 'cpu' device.")
```
